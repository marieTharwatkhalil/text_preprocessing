```python
import bs4 as bs
import urllib.request
```

```python
url=urllib.request.urlopen("https://en.wikipedia.org/wiki/France").read()
```

```python
soup=bs.BeautifulSoup(url,"lxml")
```

```python
soup
```

```python
soup.title.string
```

```python
soup.find_all("p")
```

```python
prg=soup.body.find_all("p")
len(prg)
```

```python
import re
```

```python
france=""
for parag in prg:
    france=france+parag.text
#    
france = re.sub('\W+',' ',france)
france
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
```

```python
france_tokens=word_tokenize(france)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stop_words = stopwords.words("english")
```

```python
france_filtered = []

for token in france_tokens:
    if token not in stop_words:
        france_filtered.append(token)
```

```python
france_filtered
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
france_stem = []
for word in france_filtered:
    france_stem.append(stemmer.stem(word))
```

```python
france_stem
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
france_lemma = []
for word in france_filtered:
    france_lemma.append(lemmatizer.lemmatize(word))
france_lemma    
```
